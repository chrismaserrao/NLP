```python
#plot distribution of stopwords using a barplot.
from nltk.corpus import stopwords
from collections import Counter
import matplotlib.pyplot as plt

nltk.download('stopwords')
def plot_stop(para):
    stop_words = list(stopwords.words('english'))
    print(stop_words)

    stop_tokens = Counter(word for word in word1 if word in stop_words)

    # Plot
    plt.figure(figsize=(12, 6))
    plt.bar(stop_tokens.keys(), stop_tokens.values(), color="skyblue")
    plt.xticks(rotation=90)
    plt.xlabel("Stop Words")
    plt.ylabel("Frequency")
    plt.title("Distribution of Stop Words in Text")
    plt.show()

plot_stop(para)
```

```python
#d. Remove the stopwords from the text.
def remove_stop(para):
    stop_words = set(stopwords.words("english"))
    filtered_text = " ".join(word for word in word1 if word not in stop_words)
    return filtered_text

no_stop_text = remove_stop(para)
print("text without the stop words are:", no_stop_text)
```

```python
#Determine the number of unique lemma available in the text.
nltk.download("wordnet")
from nltk.stem import WordNetLemmatizer

def uni_lemma(para):
    lemmatizer = WordNetLemmatizer()
    words = word_tokenize(para.lower())
    lemma = {lemmatizer.lemmatize(word) for word in words}
    return len(lemma)

print("no of unique lemmas:", uni_lemma(para))
```

```python
#Find all dates and convert them to the DD-MM-YYYY format.
import re
from dateutil import parser

def convert_dates(text):
    # Regex pattern to capture different date formats
    date_pattern = r'\b(\d{1,2}[-/ ]?[A-Za-z]+[-/ ]?\d{2,4}|\b[A-Za-z]+[-/ ]?\d{1,2}[-/ ]?\d{2,4})\b'

    def replace_date(match):
        try:
            parsed_date = parser.parse(match.group())  # Parse the date
            return parsed_date.strftime("%d-%m-%Y")  # Convert to DD-MM-YYYY format
        except:
            return match.group()  # Return original if parsing fails

    # Replace all dates in text
    updated_text = re.sub(date_pattern, replace_date, text)
    return updated_text

# Example text
#text = """Born on 27-March-2000, I have dedicated my research to enhancing AI-driven surgical monitoring.
          #Another breakthrough was achieved on March 15, 2021, and further tests ran on 03/04/1999."""

# Convert dates
updated_text = convert_dates(para)
print(updated_text)

```

```python
#Plot a distribution of the different values of year occurring in the text.
import re
import matplotlib.pyplot as plt
from collections import Counter

def plot_year_distribution(para):
    # Regex pattern to extract years (supports YYYY and two-digit years)
    year_pattern = r'\b(19|20)\d{2}\b'  # Matches years from 1900-2099
    
    # Find all years in the text
    years = re.findall(year_pattern, para)
    
    if not years:
        print("No years found in the text.")
        return
    
    # Count occurrences of each year
    year_counts = Counter(years)

    # Plot the distribution
    plt.figure(figsize=(8, 5))
    plt.bar(year_counts.keys(), year_counts.values(), color="skyblue")
    plt.xlabel("Year")
    plt.ylabel("Frequency")
    plt.title("Distribution of Years in Text")
    plt.xticks(rotation=45)
    plt.show()
    

# Plot the year distribution
plot_year_distribution(para)
```

```python
#k. Determine whether the text contains any phone numbers in it. 
#Ensure the phone numbers are valid if any. 
#If an invalid phone number is found, remove it from the text.
import re

def validate_and_remove_phone_numbers(para):
    # Regex pattern for common phone number formats
    phone_pattern = r'\+?\d{1,3}[-.\s]?\(?\d{3}\)?[-.\s]?\d{3}[-.\s]?\d{4}'

    # Find all potential phone numbers
    phone_numbers = re.findall(phone_pattern, para)

    valid_phone_numbers = []
    for number in phone_numbers:
        # Remove special characters for validation
        cleaned_number = re.sub(r'[-.\s()]', '', number)
        
        # Check if it's a valid 10-digit or international (11-15 digits) phone number
        if 10 <= len(cleaned_number) <= 15:
            valid_phone_numbers.append(number)
        else:
            # Remove invalid phone number from text
            text = text.replace(number, "")

    # Display results
    if valid_phone_numbers:
        print(f"Valid phone numbers found: {valid_phone_numbers}")
    else:
        print("No valid phone numbers found.")

    return text.strip()



# Process the text
cleaned_text = validate_and_remove_phone_numbers(para)
print("\nCleaned Text:\n", cleaned_text)
```
